```python

# Monte Hall (Code)<br>
''' The Monty Hall problem is a counter-intuitive statistics puzzle:

    There are 3 doors, behind which are two goats and a car.
    You pick a door (call it door A). You’re hoping for the car of course.
    Monty Hall, the game show host, examines the other doors (B & C) and opens one with a goat. 
    (If both doors have goats, he picks randomly.)

Here’s the game:
    Do you stick with door A (original guess) or switch to the unopened door? Does it matter?
    Surprisingly, the odds aren’t 50-50. If you switch doors you’ll win 2/3 of the time!'''

#List for door and goats door
doors=[0]*3
goat_doors=[0]*2
#counter to track wether to swap or not
swap=0
dont_swap=0
# track
track=0
while(track<3): #looping till 3 , ends when 3 rounds complete
#Importing random variables from range 0,2 as 3 will count as 3 in index
    import random
    x=random.randint(0,2) # xth door will have bmw
#doors configuaration
    doors[x]="BMW" 
#goat door
    for i in range(0,3): # 0,3 is taken in account as it will exclude 3rd door as it will be 4th door
        if (i==x): # if given door is bmw(xth) door
            continue   # go to the strart of loop and increment i
        else:
            doors[i]="goat" #door at ith position is goat door
            goat_doors.append(i) #so append door to particular index i here
# let user input the choice
    choice=int(input("Please Enter Your Choice: "))
# open the door
    door_open=random.choice(goat_doors) # here only those doors will open which has a goat , if it has prize game will start again
#door opened should not be the same made by the participants
    while(door_open==choice):
        door_open=random.choice(goat_doors)
#after 1st time door opens , user given choice to swap   
    choice1=input("Do You Want To Swap Y/N: ")
# if player swaps and doors are same as goat door then swap it as one goat door is already opened and 2nd one has a goat , but if player sap then he will get BMW
    if(choice1=="Y"):
        if (doors[choice]=='Goat'):
            print("Congrats You Won") # he wins
            swap=swap+1 # swap counter is incremented
        else: # if player selected BMW and swap he looses
            print("Sorry Better Luck Next Time!") 
    else:# if the player chooses not to swap and he still continues with the goat he loses if not he swaps at second one he wins
        if(doors[choice=='Goat']):
            print("You Lost")
        else:
            print("Congrats, You Won!")
            dont_swap=dont_swap+1
            
    track=track+1    #increment the loop to end
    
print(swap) # print no of swap wins
print(dont_swap) #print no of don't swap wins
```

# Example for Bubble sort:

```python
a=list(map(int,input("Enter a list of numbers to bubble sort:").split()))
def bubble(a):
    n=len(a)
    for i in range(n):
        for j in range(0,n-i-1):
            if(a[j]>a[j+1]):
                tmp=a[j]
                a[j]=a[j+1]
                a[j+1]=tmp

bubble(a)
for i in a:
    print(i)
```

```python
import numpy 
board=numpy.array{[['-','-','-'],['-','-','-'],['-','-','-']]}
p1s='X'
p2s='O'

def check_rows(symbol):
    for r in range(3):
        count=0
        for c in range(3):
            if board[r][c]==symbol:
                count =count+1
        if count==3:
           print(symbol,'won')
           return True 
    return False 
def check_cols(symbol):
    for c in range(3):
        count=0
        for r in range(3):
            if board[r][c]==symbol:
                count =count+1
        if count==3:
           print(symbol,'won')
           return True 
    return False 
def check_diagonals(symbol):
    if board[0][2]==board[1][1] and board[1][1]==board[2][0]
        print(symbol,'won')
    if board[0][0]==board[1][1] and board[1][1]==board[2][2]
        print(symbol,'won')
        return True 
    return False
def won(symbol):
    return check_rows(symbol) or check_cols(symbol) or check_diagonals(symbol)
def place(symbol):
    print(numpy.matrix(board))
    while(1):
        row=int(input('Enter row -1 or 2 or 3:'))
        col=int(input('Enter column -1 or 2 or 3:'))
        if row>0 and row<4 and col>0 and col<4 and board[row-1][col-1]=='-':
           break
        else:
           print('Invalid input,please try again with valid one')
    board[row-1][col-1] = symbol
def play():
    for turn in range(9):
        if turn%2==0:
           print('X turn')
           place(p1s)
           if won(p1s):
               break
        else:
           print('O turn')
           place(p2s)
           if won(p2s):
               break
    if not(won(p1s)) and not(won(p2s)):
        print('Draw')
play()
```

# Factorial


```python
# Python program to find the factorial of a number using recursion
def recur_fact(n):
   if n == 1:
       return n
   else:
       return n*recur_fact(n-1)
number=int(input("Enter a number : "))

if number < 0:# check is the number is negative
   print("Sorry, factorial does not exist for negative numbers")
elif number == 0:# check is the number equal to 1
   print("The factorial of 0 is 1")
else:#Recursively call the recur_fact() method
   print("The factorial of",number,"is",recur_fact(number))
```

```python
# Flames

#dependencies
import string

#defining remove matching letters fn
def remove_matching_letters(l1,l2):
    for i in range(len(l1)):
        for j in range(len(l2)):
            if l1[i]==l2[j]:
                c=l1[i]
                l1.remove(c)
                l2.remove(c)
                l=l1+['*']+l2
                return[l,True]
    l=l1+['*']+l2
    return[l,False]
    
#User Input Of Name

p1=input("Please Enter First Person Name: ")
p1=p1.lower()
p2=p1.replace(" ","")
p2=input("Please Enter Second Person Name: ")
p2=p2.lower()
p2=p2.replace(" ","")

#all letters to list

l1=list(p1)
l2=list(p2)

#flags for true or false , complete and then go fr secondd part
proceed=True
while proceed:
    return_list=remove_matching_letters(l1,l2)
    con_list=return_list[0]
    proceed=con_list[1]

#slicing for returning removed lettters words
    star_index=con_list.index('*')
    l1=con_list[:star_index]
    l2=con_list[star_index+1:]
    
#counting
count=len(l1)+len(l2)
result=['Friends','Love', 'Affection','Marriage','Enemy','Siblings']
while len(result)>1:
    split_index=(count%len(result))-1
    
 #if indexing goes positive after counting
    if split_index>=0:
        right=result[split_index+1:]
        left=result[:split_index]
        result=right+left
 #if negative
    else:
        result=result[:len(result)-1]

print(result[0])
```

# Random Walk Method(Page Rank)

```python
#Random Walk

#dependencies

import networkx as nx
import random
import matplotlib.pyplot as plt
import operator
#creating random graph
G=nx.gnp_random_graph(10,0.5,directed=True) #directed means direction
nx.draw(G, with_labels=True)
plt.show()

#picking a node randomly and incrementing counter
p=random.choice([i for i in range(G.number_of_nodes())]) #p is random source node
dict_count={}
for i in range(G.number_of_nodes()):
    dict_count[i]=0
dict_count[p]=dict_count[p]+1

#finding nebhioughrs of p and tranversing through them till end or sink
for i in range(1000000): #iteration
    list_n=list(G.neighbors(p))
#if x sink or not, if sing need to select new random node 
    if (len(list_n)==0):
        p=random.choice([i for i in range(G.number_of_nodes())])
        dict_count[p]=dict_count[p]+1 #increment iteration count
    else:
        p=random.choice(list_n) #select from list of nodes
        dict_count[p]=dict_count[p]+1 #increment iteration count
        
#veryfying page rank and random walk values as same
v=nx.pagerank(G)
#sorting 
sort_v=sorted(v.items(), key=operator.itemgetter(1))
sort_rw=sorted(dict_count.items(), key=operator.itemgetter(1))
#printing
print(sort_v)
print(sort_rw)       

#works only with high no of itteration
```

# Points Distribution Method (Page Rank)

```python
import networkx as nx 
import random
import matplotlib.pyplot as plt

def add_edges():
   nodes = list(G.nodes())   #please note G is defined below scroll the code.
   for s in nodes:
       for t in nodes:
           if s != t:
               r=random.random()    #found the probability
               if r <= 0.5:
                   G.add_edge(s,t)
                             #Explaination: if the probability is less than 0.5 then we will add edges. 
   return G


def assign_points(G):  #This Function returns a list containing 100s of length N,
                       # Where N is Length of nodes present in G.
   nodes = list(G.nodes())
   p = []
   for each in nodes:
       p.append(100)
   return p

def distribute_points(G,points):
   nodes=list(G.nodes())
   new_points=[]
   for i in range(len(nodes)):
       new_points.append(0)
       
   for n in nodes:
       out=list(G.out_edges(n))
       if len(out)==0:
           new_points[n]=new_points[n] + points[n]
       else:
           share=points[n]/len(out)
           for (src,tgt) in out:
               new_points[tgt] =new_points[tgt] + share
   return new_points
       
def keep_distributing(points,G):
   while(1):
       new_points= distribute_points(G,points)
       print(new_points)
       points = new_points
       stop = input("Press 0 to stop or any key to continue: ")
       if stop=='0' :
           break
   return new_points
       
def rank_by_points(points):
   d = {}
   for i in range(len(points)):
       d[i] = points[i]
   print(sorted(d.items(),key=lambda f:f[1]))      
       
#--------MAIN EXECUTION STARTS FROM HERE----------# 
G = nx.DiGraph() 
G.add_nodes_from([i for i in range(10)]) 
G = add_edges() #please note we didn't passed G,
                #But the function will still be able to use values of G as G is global.


nx.draw(G,with_labels=True) 
plt.show()


points = assign_points(G)  #The points will have N 100s, The N is total length of nodes present in G.

final_points = keep_distributing(points,G)


rank_by_points(final_points)


result =nx.pagerank(G) 
print(sorted(result.items(),key=lambda f:[1]))
```
